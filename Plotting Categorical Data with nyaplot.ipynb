## Plotting categorical variable with Categorical Data in Daru

```ruby
require 'daru'
```

### Part 1: Plotting a categorical variable vector

Calling plot function on a vector will plot a bar graph depicting all frequency of all variables

```ruby
# Code here should plot the graph below
v = Vector.new ['A', 'A', 'B', 'C', 'C', 'C', 'D', 'D', 'D', 'D', 'D'], type: category
v.plot
```

```ruby
# Actual code to plot this thing:
#===================================================
df = Daru::DataFrame.new({
  x: ['A', 'B', 'C', 'D'],
  y:  [2, 1, 3, 5] 
  })

df.plot type: :bar, x: :x, y: :y do |plot, diagram|
  plot.x_label "Categories"
  plot.y_label "Count"
  plot.yrange [0, 10]
end
#===================================================
```

## Part 2: Plotting dataframe with categorical variable

### Type 1: Numerical variable vs categorical variable

1. One can plot a scatter plot with categorical variable on x axis and numeric variable on y axis

```ruby
# Code here should plot the graph below
df = Daru::DataFrame.new({
  x: Daru::Vector.new(['A', 'A', 'A', 'B', 'B', 'C', 'C', 'C', 'C', 'C'], type: :category),
  y: [10,  12,   5,   1,   16,  1,   18,  5,   7,   10]
  })

df.plot type: :scatter, x: :x, y: :y
```

**NOTE**: In the plot below, the x axis should only show three values 'A', 'B' and 'C'.

```ruby
# Code used to plot the graph below
df = Daru::DataFrame.new({
  x: [1,   1,    1,   2,   2,   3,   3,   3,   3,    3],
  y: [10,  12,   5,   1,   16,  1,   18,  5,   7,   10]
  })

df.plot type: :scatter, x: :x, y: :y do |plot, diagram|
  plot.x_label "x"
  plot.y_label "y"
  plot.yrange [0, 20]
  plot.xrange [0, 4]
end
```

In a strip plot like above, the scatterplot points will usually overlap. This makes it difficult to see the full distribution of data. There will be an option **jitter** to adjust the positions (only along the categorical axis) using some random “jitter”.

## Type 2: Different diagrams for different categories

a. Multiple scatterplots for different categories

```ruby
# Code here should plot the graph below
df = Daru::DataFrame.new({
  cat: ['a', 'a', 'b', 'b', 'b', 'b', 'c', 'c', 'c', 'c', 'c'],
  x:   [1,   2,  6,   8,   9,   3,   5,   6,  10,   5,   9],
  y:   [4,   6,  5,   2,   6,   2,   6,   10,  9,   1,   9]
  })

df.plot type: :scatter, categorized_by: :cat, categorization_method: :shape
```

```ruby
# Code used to plot the graph below
df = Daru::DataFrame.new({
  cat: ['a', 'a', 'b', 'b', 'b', 'b', 'c', 'c', 'c', 'c', 'c'],
  x:   [1,   2,  6,   8,   9,   3,   5,   6,  10,   5,   9],
  y:   [4,   6,  5,   2,   6,   2,   6,   10,  9,   1,   9]
  })

df.plot(type: :scatter, x: :x, y: :y) do |plot, diagram|
  plot.yrange [0, 15]
  plot.xrange [0, 15]
  diagram.color(Nyaplot::Colors.qual)
  diagram.shape_by(:cat)
end
```

One can also categorize the different plots by color

```ruby
# Code here should plot the graph below
df.plot type: :scatter, categorized_by: :cat, categorization_method: :color
```

```ruby
# Code used to plot the graph below
df = Daru::DataFrame.new({
  cat: ['a', 'a', 'b', 'b', 'b', 'b', 'c', 'c', 'c', 'c', 'c'],
  x:   [1,   2,  6,   8,   9,   3,   5,   6,  10,   5,   9],
  y:   [4,   6,  5,   2,   6,   2,   6,   10,  9,   1,   9]
  })

df.plot(type: :scatter, x: :x, y: :y) do |plot, diagram|
  plot.yrange [0, 15]
  plot.xrange [0, 15]
  diagram.color(Nyaplot::Colors.qual)
  diagram.fill_by(:cat)
end
```
